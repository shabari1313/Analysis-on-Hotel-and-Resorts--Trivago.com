### Karnataka

```python
price=[]
rate=[]
title=[]
loc=[]
deals=[]
feed=[]
other=[]
state=[]
driver = webdriver.Chrome()

for page_num in range(1, 37): 
    url = f"https://www.trivago.in/en-IN/srl/hotels-karnataka-india?search=200-64938;dr-20250919-20250924-s;pa-{page_num};rc-1-2"
    driver.get(url)
    
    WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.TAG_NAME, "h2"))
    )
        
    soup = BeautifulSoup(driver.page_source, "html.parser")
        
    for i in soup.find_all('div', class_='_96iWr8'):
        t=i.find('h2',class_='tODn3R _5NF9ws KyLbnf')
        title.append(t.text)

        p=i.find('span',class_='SmLAfp Eqg9Bk Dtphjk EXnXoV')
        price.append(p)

        r=i.find('div',class_='NFI0QA')
        rate.append(r.text)

        l=i.find('span',class_="wcZPz9")
        loc.append(l.text)

        d=i.find('p',class_="vBQGhG")
        deals.append(d)
        
        f=i.find('span',class_='_1Wh5Tf')
        feed.append(f.text)

        o=i.find('button',class_='_6Ey4vS _44WTUH')
        if o:
            other.append(o)
        else:
            other.append(np.nan)
        state.append('Karnataka')

driver.quit()
print(len(title))
print(len(price))
print(len(rate))
print(len(loc))
print(len(deals))
print(len(feed))
print(len(other))
```

```python
df1 = pd.DataFrame({'Hotel_Name':title,'Price':price,'Ratings':rate,'City':loc,'Deals':deals,'Feedback':feed,'Other_sites':other,'State':state})
df1.head()
```
